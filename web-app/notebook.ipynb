```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression
import pickle

# 1. Load the data (Make sure you have a folder named 'data' with 'ufos.csv' inside!)
ufos = pd.read_csv('./data/ufos.csv')

# 2. Clean and format the dataframe
ufos = pd.DataFrame({'Seconds': ufos['duration (seconds)'], 
                     'Country': ufos['country'],
                     'Latitude': ufos['latitude'],
                     'Longitude': ufos['longitude']})

# 3. Drop nulls and filter for sightings between 1 and 60 seconds
ufos.dropna(inplace=True)
ufos = ufos[(ufos['Seconds'] >= 1) & (ufos['Seconds'] <= 60)]

# 4. Convert the text 'Country' into numbers so the AI can read it
ufos['Country'] = LabelEncoder().fit_transform(ufos['Country'])

# 5. Define our Features (X) and Target (y)
Selected_features = ['Seconds','Latitude','Longitude']
X = ufos[Selected_features]
y = ufos['Country']

# 6. Split the data into Training and Testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# 7. Train the Logistic Regression Model
model = LogisticRegression(max_iter=500) # Added max_iter to prevent convergence warnings
model.fit(X_train, y_train)

# 8. Check the accuracy
predictions = model.predict(X_test)
print('Accuracy: ', accuracy_score(y_test, predictions))

# 9. "Pickle" (save) the trained model for our web app
model_filename = 'ufo-model.pkl'
pickle.dump(model, open(model_filename,'wb'))
print(f"\nModel saved successfully as {model_filename}!")
```
